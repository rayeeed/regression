```python
import pandas as pd


df1 = pd.read_excel("/content/drive/MyDrive/datasets_Rayeed_GSV_POI_LU/model_49_ID.xlsx")

df2 = pd.read_excel("/content/drive/MyDrive/datasets_Rayeed_GSV_POI_LU/combined_total_v3.0.xlsx")

# Select rows from the second DataFrame based on the values in the "OID_" column
df_selected = df2[df2['OID_'].isin(df1['OID'])]
```

```python
df1
```

```python
df_selected 
```

```python
df_selected_final= df1.merge(df_selected, left_on='OID',right_on='OID_')
```

```python
df_selected_final
```

```python
df_selected_final.drop(['OID','latitude','longitude'],inplace=True,axis=1)
```

```python
df_selected_final
```

```python
# Output the selected rows to a new Excel file
df_selected_final.to_excel("/content/drive/MyDrive/datasets_Rayeed_GSV_POI_LU/model_49.xlsx", index=False)
```
